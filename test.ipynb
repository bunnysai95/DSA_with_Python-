```python
# Binary search 
nums = [4,5,6,7,0,1,2]
target = 0

l, r = 0, len(nums) -1
while l <= r:
    mid = (l+r) // 2
    if target == nums[mid]:
        print(mid)
    # left half
    if nums[l] <= nums[mid]:
        if target > nums[mid] or target < nums[l]:
            l = mid+1 # go right
        else:
            r = mid-1  # go left
    # right half is sorted
    else:
        if target < nums[mid] or target > nums[r]:
            r = mid -1 # go left
        else:
            l = mid+1 # go right
    # print("-1")
```
